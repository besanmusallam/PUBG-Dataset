###distribution plot

```python
numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns

num_cols = 3
num_features = len(numeric_cols)

fig, axes = plt.subplots(num_features // num_cols + 1, num_cols, figsize=(15, 15))
axes = axes.flatten()

colors = sns.color_palette('muted', num_features)

for i, (col, color) in enumerate(zip(numeric_cols, colors)):
    sns.histplot(data[col], kde=True, ax=axes[i], color=color)
    axes[i].set_title(f'Distribution of {col}')

for j in range(num_features, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

### the data is rather a uniform distribution than a normal distribution
diabetes, Family History, Smoking, Obesity, Alcohol Consumption ,Medication Use , Previous Heart Problems, Heart Attack Risk are nominal data

# Numerical Data

###OUTLIERS

```python
feature_names = ['Age', 'Cholesterol', 'Heart Rate', 'Exercise Hours Per Week', 'Stress Level',
                 'Sedentary Hours Per Day', 'Income', 'BMI', 'Triglycerides',
                 'Physical Activity Days Per Week', 'Sleep Hours Per Day']
num_plots = len(feature_names)
num_cols = 4
num_rows = -(-num_plots // num_cols)

fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, 5*num_rows))
axes = axes.flatten()
custom_palette = sns.color_palette("husl", num_plots)

for i, (feature, color) in enumerate(zip(feature_names, custom_palette)):
    sns.boxplot(y=data[feature], ax=axes[i], color=color)
    axes[i].set_title(f'Boxplot of {feature}', fontsize=14)
    axes[i].set_ylabel('')
    axes[i].set_xlabel('')
    axes[i].tick_params(axis='x', rotation=45)

for j in range(num_plots, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

The uniform distribution of these features indicates the absence of outliers

###correlation (numeric features, target)

```python
corr_matrix = data.corr()
corr_matrix
```

```python
# Plot the correlation matrix using heatmap
fig, ax = plt.subplots(figsize=(15,10))    #size
sns.heatmap(corr_matrix, ax=ax, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

There are no strong relationships observed either between the features themselves or between the features and the target variable

# Categorical Data

```python
data['Sex'].value_counts()
```

```python
female_count = (data['Sex'] == 'Female').sum()
male_count = (data['Sex'] == 'Male').sum()

labels = ['Female', 'Male']

colors = ['salmon', 'pink']

sizes = [female_count, male_count]

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

plt.axis('equal')

plt.title('Gender Distribution')

plt.show()
```

###number of smokers for each country

```python
heart_attack_data = data[data['Heart Attack Risk'] == 1]
country_counts = heart_attack_data['Country'].value_counts()
colors = sns.color_palette("muted", len(country_counts))
plt.figure(figsize=(10, 6))
sns.countplot(x='Country', data=heart_attack_data, palette=colors, order=country_counts.index)
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Number of Heart Attacks by Country')
plt.show()
```
